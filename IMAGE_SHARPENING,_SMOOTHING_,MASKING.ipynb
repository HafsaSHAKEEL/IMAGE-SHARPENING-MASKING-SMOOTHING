**TASK 1**

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the image
image = cv2.imread('Stars-1.jpg')
```

**Image Averaging**

```python
# Image Averaging for Noise Reduction
kernel_size = (5, 5)
image_avg = cv2.blur(image, kernel_size)
cv2_imshow(image_avg)
```

**Brightness Change**

```python
# Changing of Brightness
alpha = 1.5  # Brightness factor
beta = 30    # Brightness offset
image_bright = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
cv2_imshow(image_bright)
```

**Mask**

```python
# Removal of Background from Images (Assuming you have a mask image)
mask = cv2.imread('mask.jpg', 0)
image_bg_removed = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(image_bg_removed)
```

**Image Negative**

```python
# Image Negative
image_negative = 255 - image
cv2_imshow(image_negative)
```

**Point Operation**

```python
# Point Operations (Intensity Transformation)
gamma = 0.5  # Gamma correction factor
image_gamma = np.uint8(cv2.pow(image / 255.0, gamma) * 255)
cv2_imshow(image_gamma)
```

**Histogram**

```python
# Histogram based processing
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(image_gray)
cv2_imshow(equalized_image)
```

**Mask Processing**

```python
# Mask Processing (Assuming you have a mask image)
mask = cv2.imread('mask.jpg', 0)
masked_image = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(masked_image)
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the images
image1 = cv2.imread('Stars-1.jpg')
image2 = cv2.imread('1.jpg')

# Resize images to ensure they have the same dimensions
image1_resized = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

# Calculate Mean Squared Error (MSE)
mse = np.mean((image1_resized - image2) ** 2)

# Display the results
print(f"Mean Squared Error (MSE): {mse}")
```

**TASK 3**

**Sharpening**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the original image
original_image = cv2.imread('Stars-1.jpg')

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(original_image, (0,0), 3)

# Calculate the sharpened image using unsharp masking
sharpened_image = cv2.addWeighted(original_image, 1.5, blurred_image, -0.5, 0)

# Calculate the improvement in visibility
improvement_percentage = ((np.sum(sharpened_image) - np.sum(original_image)) / np.sum(original_image)) * 100

# Display the original and sharpened images
cv2_imshow(original_image)
cv2_imshow(sharpened_image)

# Print the improvement percentage
print(f"Improvement in visibility: {improvement_percentage:.2f}%")
```

**TASK 4**

**Smoothening**

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the original image
original_image = cv2.imread('Stars-1.jpg')

# Apply Gaussian blur to smoothen the image
smoothed_image = cv2.GaussianBlur(original_image, (15, 15), 0)

# Apply a sharpening filter to recover the original image
recovered_image = cv2.addWeighted(original_image, 1.5, smoothed_image, -0.5, 0)

# Display the original and recovered images
cv2_imshow(original_image)
cv2_imshow(recovered_image)
```
